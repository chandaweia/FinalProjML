```python
def auto_text(rects):
    for rect in rects:
        ax.text(rect.get_x(), rect.get_height(), rect.get_height(), ha='left', va='bottom')
def auto_label(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height), # put the detail data
                    xy=(rect.get_x() + rect.get_width() / 2, height), # get the center location.
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
x = data1['ID']
y = data1['MNIST']
width = 0.5

#x = np.arange(len(x))
fig,ax = plt.subplots()
weight = ax.bar(x,y,width,color='coral')
#ax.set_xticks(x + width/2)#将坐标设置在指定位置
ax.set_xticks(x)
ax.set_xticklabels(x)#将横坐标替换成
#ax.set_xlabel('')
#ax.set_ylabel('Number')
ax.set_title('Distribution of MNIST')
auto_text(weight)

plt.legend()
```
